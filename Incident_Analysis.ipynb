```python
#droping coulmns that are not rrequired
df = df[['number',
         'opened_at',
         'resolved_at',
         'impact',
         'category',
         'assignment_group',
         'closed_code',
         'incident_state']]
df
```

```python
#renaming the columns
df = df.rename(columns={
    'number': 'incident_id',
    'opened_at': 'created_at',
    'impact': 'severity',
    'assignment_group': 'team_assigned',
    'close_code': 'root_cause',
    'incident_state': 'status'
})
df
```

```python
df['created_at'] = pd.to_datetime(df['created_at'])
```

```python
df['shift'] = df['created_at'].dt.hour.apply(
    lambda x: 'Night' if x < 8 else ('Morning' if x < 16 else 'Evening')
)
```

```python
root_cause_map = {
    "code 1": "Network Failure",
    "code 2": "Database Issue",
    "code 3": "Application Bug",
    "code 4": "Configuration Error",
    "code 5": "Server Overload",
    "code 6": "Authentication Failure",
    "code 7": "API Timeout",
    "code 8": "Memory Leak",
    "code 9": "Disk Space Issue",
    "code 10": "Third-Party Service Failure",
    "code 11": "Data Corruption",
    "code 12": "Security Policy Block",
    "code 13": "Job Scheduling Failure",
    "code 14": "Integration Failure",
    "code 15": "Deployment Error",
    "code 16": "Hardware Failure",
    "code 17": "Unknown System Error",
    "Unknown": "Not Documented"
}

res_inc_df["root_cause"] = res_inc_df["closed_code"].map(root_cause_map)
```

```python
res_inc_df= res_inc_df.drop(columns= 'closed_code')
```

```python
res_inc_df= res_inc_df.drop(columns= 'status')
res_inc_df
```

```python
# extracting numbers from category, to convert them into meaningful values.

res_inc_df["category_num"] = res_inc_df["category"].str.extract(r'(\d+)')
res_inc_df["category_num"] = pd.to_numeric(res_inc_df["category_num"], errors="coerce")
```

```python
#Mapping the values using conditions
def map_category(x):
    if pd.isna(x):
        return "Unknown"
    elif 1 <= x <= 10:
        return "Network Issues"
    elif 11 <= x <= 20:
        return "Database Issues"
    elif 21 <= x <= 30:
        return "Application Errors"
    elif 31 <= x <= 40:
        return "Server Infrastructure"
    elif 41 <= x <= 50:
        return "Integration / API Issues"
    elif 51 <= x <= 60:
        return "Security / Access Issues"
    else:
        return "Monitoring / Alerting"

res_inc_df["category_clean"] = res_inc_df["category_num"].apply(map_category)
```

```python
res_inc_df["category_clean"].value_counts()
```

```python
res_inc_df= res_inc_df.drop(columns= 'category')
res_inc_df= res_inc_df.drop(columns= 'category_num')
res_inc_df
```

```python
#assigning the team based on the category
def assign_team(category):
    
    if category == "Network Issues":
        return "Network Operations Team"
    
    elif category == "Database Issues":
        return "Database Administration Team"
    
    elif category == "Application Errors":
        return "Application Support Team"
    
    elif category == "Server Infrastructure":
        return "Infrastructure Operations Team"
    
    elif category == "Integration / API Issues":
        return "Integration Support Team"
    
    elif category == "Security / Access Issues":
        return "Security Operations Team"
    
    elif category == "Monitoring / Alerting":
        return "Monitoring & DevOps Team"
    
    else:
        return "Service Desk Team"


res_inc_df["assigned_team"] = res_inc_df["category_clean"].apply(assign_team)
```

```python
res_inc_df= res_inc_df.drop(columns= 'team_assigned')
res_inc_df
```
